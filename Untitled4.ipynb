```python
def load_nii(path):
    return nib.load(path).get_fdata()

base_path = "/content/drive/MyDrive/brats2021/"
flair = load_nii(base_path + "BraTS2021_00495_flair.nii.gz")
t1    = load_nii(base_path + "BraTS2021_00495_t1.nii.gz")
t1ce  = load_nii(base_path + "BraTS2021_00495_t1ce.nii.gz")
t2    = load_nii(base_path + "BraTS2021_00495_t2.nii.gz")
mask  = load_nii(base_path + "BraTS2021_00495_seg.nii.gz")
```

```python
image = np.stack([flair, t1, t1ce, t2], axis=-1)
def normalize(volume):
    nonzero = volume[volume > 0]
    mean, std = nonzero.mean(), nonzero.std()
    volume = (volume - mean) / std
    volume = np.clip(volume, -5, 5)
    return volume

for i in range(4):
    image[..., i] = normalize(image[..., i])
```

```python
def extract_tumor_slices(img, msk):
    X, Y = [], []
    for i in range(msk.shape[2]):
        if np.any(msk[:, :, i] > 0):
            X.append(img[:, :, i, :])
            Y.append(msk[:, :, i])
    return np.array(X), np.array(Y)

X, Y = extract_tumor_slices(image, mask)
```

```python
import cv2

def resize_data(X, Y, size=128):
    Xr, Yr = [], []
    for i in range(len(X)):
        Xr.append(cv2.resize(X[i], (size, size)))
        Yr.append(cv2.resize(Y[i], (size, size), interpolation=cv2.INTER_NEAREST))
    return np.array(Xr), np.array(Yr)

X, Y = resize_data(X, Y)
Y[Y == 4] = 3
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
import tensorflow as tf
from tensorflow.keras import backend as K

NUM_CLASSES = 4

def dice_coef(y_true, y_pred, smooth=1e-6):

    y_true = tf.one_hot(tf.cast(y_true, tf.int32), NUM_CLASSES)


    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)

    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (
        K.sum(y_true_f) + K.sum(y_pred_f) + smooth
    )


def iou(y_true, y_pred, smooth=1e-6):
    y_true = tf.one_hot(tf.cast(y_true, tf.int32), NUM_CLASSES)

    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)

    intersection = K.sum(y_true_f * y_pred_f)
    union = K.sum(y_true_f) + K.sum(y_pred_f) - intersection

    return (intersection + smooth) / (union + smooth)
```

```python
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model

def unet(input_shape=(128,128,4), num_classes=4):
    inputs = Input(input_shape)

    c1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    c1 = Conv2D(64, 3, activation='relu', padding='same')(c1)
    p1 = MaxPooling2D()(c1)

    c2 = Conv2D(128, 3, activation='relu', padding='same')(p1)
    c2 = Conv2D(128, 3, activation='relu', padding='same')(c2)
    p2 = MaxPooling2D()(c2)

    c3 = Conv2D(256, 3, activation='relu', padding='same')(p2)
    c3 = Conv2D(256, 3, activation='relu', padding='same')(c3)

    u4 = UpSampling2D()(c3)
    u4 = concatenate([u4, c2])
    c4 = Conv2D(128, 3, activation='relu', padding='same')(u4)

    u5 = UpSampling2D()(c4)
    u5 = concatenate([u5, c1])
    c5 = Conv2D(64, 3, activation='relu', padding='same')(u5)

    outputs = Conv2D(num_classes, 1, activation='softmax')(c5)

    return Model(inputs, outputs)

model = unet()
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=[dice_coef, iou]
)

model.summary()
```

```python
history = model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=20,
    batch_size=8
)
```

```python
loss, dice, iou_score = model.evaluate(X_val, Y_val)
print("Dice Score:", dice)
print("IoU Score:", iou_score)
```

```python
def predict_brain_tumor_for_patient(
    model,
    flair_path,
    t1_path,
    t1ce_path,
    t2_path,
    slice_index=None,
    img_size=128
):

    def load_nii(path):
        return nib.load(path).get_fdata()

    def normalize(volume):
        nonzero = volume[volume > 0]
        mean, std = nonzero.mean(), nonzero.std()
        volume = (volume - mean) / std
        return np.clip(volume, -5, 5)


    flair = load_nii(flair_path)
    t1    = load_nii(t1_path)
    t1ce  = load_nii(t1ce_path)
    t2    = load_nii(t2_path)


    image = np.stack([flair, t1, t1ce, t2], axis=-1)


    for c in range(4):
        image[..., c] = normalize(image[..., c])

    slices = []
    for i in range(image.shape[2]):
        resized = cv2.resize(image[:, :, i, :], (img_size, img_size))
        slices.append(resized)

    X = np.array(slices)


    preds = model.predict(X, verbose=0)
    pred_masks = np.argmax(preds, axis=-1)


    if slice_index is None:
        slice_index = image.shape[2] // 2

    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.title("FLAIR")
    plt.imshow(X[slice_index][:,:,0], cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Predicted Segmentation")
    plt.imshow(pred_masks[slice_index], cmap='jet')
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Overlay")
    plt.imshow(X[slice_index][:,:,0], cmap='gray')
    plt.imshow(pred_masks[slice_index], cmap='jet', alpha=0.4)
    plt.axis("off")

    plt.show()

    return pred_masks
```

```python
import kagglehub
import os
from collections import defaultdict
import nibabel as nib
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Download latest version
path = kagglehub.dataset_download("awsaf49/brats2020-training-data")

print("Path to dataset files:", path)

# Construct the full path to the actual training data directory
# Based on common Kaggle dataset structure for BraTS 2020
actual_data_path = os.path.join(path, "MICCAI_BraTS2020_TrainingData")

def organize_brats_patients(dataset_base_path):
    patients = defaultdict(dict)
    if not os.path.exists(dataset_base_path):
        print(f"Warning: Dataset path '{dataset_base_path}' does not exist.")
        return patients

    for root, dirs, files in os.walk(dataset_base_path):
        # We are looking for directories that directly contain the NIfTI files for a patient.
        # These directories usually start with 'BraTS20_Training_'.
        patient_id_candidate = os.path.basename(root)

        # Check if the current directory is a patient directory and it contains NIfTI files
        # The 'seg' file is usually a good indicator of a complete patient scan folder
        nifti_files_in_dir = [f for f in files if f.endswith(".nii.gz")]
        if nifti_files_in_dir and patient_id_candidate.startswith(('BraTS20_Training_', 'BraTS20_')):
            for file in nifti_files_in_dir:
                full_path = os.path.join(root, file)
                modality_raw = file.replace(".nii.gz", "")

                # Extract modality from filename, which is the part after patient_id_
                # Example: BraTS20_Training_001_flair.nii.gz -> flair
                if modality_raw.startswith(patient_id_candidate + '_'):
                    modality_name = modality_raw.replace(patient_id_candidate + '_', '')
                else:
                    # Fallback for other naming conventions if any
                    parts_of_filename = modality_raw.split('_')
                    if len(parts_of_filename) > 1:
                        modality_name = parts_of_filename[-1]
                    else:
                        modality_name = modality_raw # Use raw name if no clear patient_id prefix

                # Standardize modality names
                modality_map = {
                    'flair': 'flair',
                    't1': 't1',
                    't1ce': 't1ce',
                    't2': 't2',
                    'seg': 'seg'
                }
                modality = modality_map.get(modality_name.lower(), modality_name.lower())

                patients[patient_id_candidate][modality] = full_path
    return patients

patients = organize_brats_patients(actual_data_path)
print("Number of patients:", len(patients))

def load_nifti(path):
    return nib.load(path).get_fdata()

def prepare_patient_input(patient_files, img_size=128):
    volumes = []
    required_modalities = ["flair", "t1", "t1ce", "t2"]

    # Get the patient_id for error messages
    patient_id_for_error = "Unknown Patient"
    # Find patient_id from the first file path if available
    if patient_files:
        first_file_path = next(iter(patient_files.values()))
        # Extract patient_id from the folder name
        patient_id_for_error = os.path.basename(os.path.dirname(first_file_path))


    for modality in required_modalities:
        if modality not in patient_files:
            raise ValueError(f"Missing modality '{modality}' for patient: {patient_id_for_error}")

        vol = load_nifti(patient_files[modality])

        # Apply the same Z-score normalization and clipping as during model training (from cell _n2b1dz6CYBE)
        nonzero = vol[vol > 0]
        if nonzero.size > 0:
            mean, std = nonzero.mean(), nonzero.std()
            normalized_vol = (vol - mean) / (std + 1e-8)
        else:
            normalized_vol = vol
        normalized_vol = np.clip(normalized_vol, -5, 5)

        volumes.append(normalized_vol)

    stacked = np.stack(volumes, axis=-1)
    return stacked

def volume_to_slices(volume, target_size=128):
    slices = []
    depth = volume.shape[2]

    for i in range(depth):
        slice_ = volume[:, :, i, :]
        resized_slice = np.zeros((target_size, target_size, volume.shape[-1]), dtype=slice_.dtype)
        for c in range(volume.shape[-1]):
            resized_slice[:, :, c] = cv2.resize(slice_[:, :, c],
                                           (target_size, target_size),
                                           interpolation=cv2.INTER_LINEAR)

        slices.append(resized_slice)

    return np.array(slices)

def predict_patient(model, patient_files):
    volume = prepare_patient_input(patient_files)
    slices = volume_to_slices(volume)

    preds = model.predict(slices, verbose=0)
    preds = np.argmax(preds, axis=-1)

    return preds

# --- Main execution block for prediction and visualization ---
if patients:
    patient_id = list(patients.keys())[0]
    print(f"Predicting for patient: {patient_id}")
    try:
        predicted_masks = predict_patient(model, patients[patient_id])
    except ValueError as e:
        print(f"Error preparing patient input for {patient_id}: {e}")
        predicted_masks = np.array([])

    def visualize_prediction(patient_files, pred_masks, slice_id=None):
        flair_path = patient_files.get("flair")
        seg_path = patient_files.get("seg")

        if not flair_path:
            print("FLAIR modality not found for visualization.")
            return

        flair_full_volume = load_nifti(flair_path)

        if slice_id is None:
            slice_id = flair_full_volume.shape[2] // 2

        slice_id = min(max(0, slice_id), flair_full_volume.shape[2] - 1)

        flair_slice = flair_full_volume[:, :, slice_id]
        flair_resized = cv2.resize(flair_slice, (128, 128), interpolation=cv2.INTER_LINEAR)

        if seg_path:
            gt_slice = load_nifti(seg_path)[:, :, slice_id]
            gt_resized = cv2.resize(gt_slice, (128, 128), interpolation=cv2.INTER_NEAREST)
            gt_title = "Ground Truth"
        else:
            gt_resized = np.zeros_like(flair_resized)
            gt_title = "No Ground Truth Available"

        if pred_masks.size == 0 or pred_masks.shape[0] == 0:
            print("No predicted masks available for visualization.")
            return

        effective_pred_slice_id = min(max(0, slice_id), pred_masks.shape[0] - 1)
        effective_pred_mask = pred_masks[effective_pred_slice_id]

        plt.figure(figsize=(12,4))
        plt.subplot(1,3,1); plt.title("FLAIR"); plt.imshow(flair_resized, cmap="gray")
        plt.subplot(1,3,2); plt.title(gt_title); plt.imshow(gt_resized)
        plt.subplot(1,3,3); plt.title("Prediction"); plt.imshow(effective_pred_mask, cmap='jet')
        plt.show()

    if 'predicted_masks' in locals() and predicted_masks.size > 0:
        first_patient_flair_path = patients[patient_id].get("flair")
        if first_patient_flair_path:
            first_patient_flair_volume = load_nifti(first_patient_flair_path)
            middle_slice_idx = first_patient_flair_volume.shape[2] // 2
            visualize_prediction(patients[patient_id], predicted_masks, slice_id=middle_slice_idx)
        else:
            print("Cannot determine slice_id for visualization: FLAIR volume not found for the first patient.")
    else:
        print("No masks predicted for visualization.")
else:
    print("No patients found in the dataset for prediction. Please check the dataset path and structure.")
```
